```python
import pandas as pd
import matplotlib.pyplot as plt

data = [['EMP001','M',34,123,'Normal',350],
        ['EMP002','F',24,124,'Normal',350],
        ['EMP003','M',39,150,'Normal',350],
        ['EMP006','M',39,160,'Normal',350],
        ['EMP007','M',39,165,'Normal',250],
        ['EMP004','M',29,110,'Obesity',150],
        ['EMP005','F',23,140,'Underweight',300],
        ['EMP005','F',23,129,'Underweight',300],]
dataframe = pandas.DataFrame(data,columns = ['Name','Gender','Age','Sales','BMI','Income'])
dataframe.plot.bar()
plt.bar(dataframe['Name'],dataframe['BMI'],dataframe['Income'])
plt.xlabel('Name')
plt.xlabel('Income')
plt.xlabel('BMI')
plt.show()
```

```python
import pandas
import matplotlib.pyplot as plt

# linewidths- width of marker border
# edgecolor- marker border color

age=[23,32,45,58,66,75]
income=[200,4500,7800,10100,15470,19343]

# x_axis_data- An array containing x-axis data
# y_axis_data- An array containing y-axis data
# s- marker size (can be scalar or array of size equal to size of x or y)

plt.scatter(income,age,c ="black",linewidths=2,marker="s",edgecolor="red",s=50)
plt.xlabel("income")
plt.ylabel("age")
plt.show()
```

```python
import matplotlib.pyplot as plt
x1 = [89, 43, 36, 36, 95, 10,
      66, 34, 38, 20]
 
y1 = [21, 46, 3, 35, 67, 95,
      53, 72, 58, 10]
plt.scatter(x1, y1, c ="black",
            linewidths = 2,
            marker ="s",
            edgecolor ="purple",
            s = 50)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
