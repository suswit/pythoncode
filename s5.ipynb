```python
import pandas
import matplotlib.pyplot as plt

# linewidths- width of marker border
# edgecolor- marker border color

age=[23,32,45,58,66,75]
income=[200,4500,7800,10100,15470,19343]

# x_axis_data- An array containing x-axis data
# y_axis_data- An array containing y-axis data
# s- marker size (can be scalar or array of size equal to size of x or y)

plt.scatter(income,age,c ="black",linewidths=2,marker="s",edgecolor="red",s=50)
plt.xlabel("income")
plt.ylabel("age")
plt.show()
```

```python
x1=[1,2,3]
y1=[2,4,1]
plt.plot(x1,y1,label = "line1")

x2=[1,2,3]
y2=[4,1,3]
plt.plot(x2,y2,label = "line2")

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Two lines on same graph!')

plt.legend()
plt.show()
```

```python
left = [1,2,3,4,5]
height = [10,24,36,40,5]

tick_label = ['one','two','three','four','five']

plt.bar(left,height,tick_label = tick_label,
        width = 0.8,color = ['green','yellow','red','blue','purple'])

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar Chart!')

# plt.legend()
plt.show()
```
